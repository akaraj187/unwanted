## Your ML Setup: The Compact Guide 
1. Your "Workshop": Use WSL (Linux)
What it is: WSL (Windows Subsystem for Linux) is a real Linux environment üêß running inside your Windows.

Why? The entire ML and data science world (servers, tools, libraries) is built to run on Linux. Using WSL makes installing tools 100x easier and more stable than on Windows.

Analogy: PowerShell is for managing your Windows house. WSL is the professional engineering workshop you build in your backyard. You do all your serious ML work in the workshop.

2. The Rule: Work Inside the Workshop
What it is: Always store your projects in the Linux file system, not the Windows one.

Fast (Good) üëç: ~/projects/my_project (This is short for /home/your_username/projects/my_project)

Slow (Bad) üëé: /mnt/c/Users/YourName/projects

Analogy: Working in ~ is fast because all your tools and materials are in the workshop. Working in /mnt/c/ is extremely slow because you have to run back to the house to grab every single screw.

3. The "Toolbox": Use a venv for Every Project
What it is: A Virtual Environment (venv) is a private, isolated "toolbox" üß∞ for each of your projects.

Why? It stops projects from breaking each other. Your ML project might need pandas 2.0, while an old web project needs pandas 1.0. A venv keeps their tools separate.

Key Commands:

Create Toolbox: python3 -m venv venv (Do this once per project).

Open Toolbox: source venv/bin/activate (Do this every time you start working).

Put Tools in your ToolBox : Installing Libraries you need like:

pip install numpy (Installs numpy only into this project's toolbox)

jupyter lab (Runs Jupyter using this toolbox's libraries)

Key Concept: You can have many toolboxes (one per project), but you only activate (open) one at a time.

4. 4. Your Daily Workflow: Putting It All Together
This is the process you'll follow every time you start a new coding session for this project.

Open your WSL (Ubuntu) terminal.

Navigate to your project's fast Linux folder: cd ~/projects/my_ml_project

Activate its specific toolbox (which already has your tools): source venv/bin/activate

Work!

Run your code (python train_model.py).

Start your tools (jupyter lab or code .).

You only run pip install again if you need to add a new tool to your toolbox.

When finished, just close the terminal. The venv deactivates automatically.
