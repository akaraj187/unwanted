1.Task: Create a matrix of zeros with shape (4, 3) (4 rows, 3 columns). Then, print its shape and data type.

```python
import numpy as np
weights=np.zeros([4,3])
print(weights)
print(weights.shape) #shape of matrix
print(weights.dtype)  #datatype of matrix elememts
print(weights.ndim) # dimension of array
print(weights.size) # size of matix (count of elements)
```

4.Create matrix X with shape (2, 3).

Create matrix Y with shape (3, 2).

Multiply them (np.dot). What is the shape of the result?

Bonus: Try multiplying Y and X (reverse order). Does it work? What is the shape?

Why: The order of multiplication changes the output dimensions. This creates "Silent Bugs."

```python
x=np.arange(6)
x=np.reshape(x,[2,3])
print('Matrix X')
print(x)

y=np.arange(6,12)
y=np.reshape(y,[3,2])
print('Matrix Y')
print(y)

print('x dot y')
w=np.dot(x,y)
print(w)
print(w.shape)

print('y dot x')
u=np.dot(y,x)
print(u)
print(u.shape)
```

Create a random array with negative numbers: arr = np.array([10, -5, 2, -1]).

The Task: Use Boolean Masking (square brackets []) to set all negative numbers to 0.

Hint: arr[arr < 0] selects the specific numbers. Can you assign 0 to them?

```python
'''arr = np.array([10, -5, 2, -1])
print(arr)

for i in range (len(arr)):
    if(arr[i]<0):
        arr[i]=0

print(arr)'''

#loop takes long so boolean masking is efficient and fast.

arr = np.array([10, -5, 2, -1])

# The "Mask" (True/False check)
# This creates: [False, True, False, True]
mask = arr < 0 

# The "Vectorized" Update
# It says: "Where the mask is True, set value to 0"
arr[mask] = 0 

print(arr)
```
